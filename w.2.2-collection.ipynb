# Collections (the complex data types)
## Collections: store all other data types (stores int, strings, booleans)
# [ ] index notations

- sets
- tuples
- list
- dictionaries
- combine collections

```python
## Sets: no duplicates, no indexed (no order), itarable (loopable), collection, mutable (chanagable)

set1 = {'Ben','Jason','Darlene','Erin'}
set_declared = set(('Ben', 'Jason','G','G','Darlene'))
print(f"set1 ={set1}")
print(f"set_declared = {set_declared}")
```

```python
## Method: function that belongs to an object ( ex... set1. would call a method thats linked to that dot)
```

```python
# Method to find overlaps
overlaps = set1.intersection(set_declared)
overlaps
```

```python
# Method to find the difference between two sets
diff = set1.difference(set_declared)
diff
```

```python
# to add a new variable into a set
set1.add('Halle')
set1
```

```python
# to remove variabke from a set
set1.remove('Darlene')
set1
```

```python
# Tuple: collection that exsist in parentheses
# non mutable (can't change), duplicates, itarable, indexed

tuple1 = (1,2,3,4)
tuple_declared = tuple((3,4,5,6))

print(f'tuple1 = {tuple}')
print(f'tuple_declared = {tuple_declared}')
```

```python
tuple1 = (1,2,3,4,5) # overwriting, not changing
tuple1
```

```python
# count occurances of a value
tuple1. count(2)
```

```python
# index: returns index of value
tuple1.index(1)
```

```python
# returns value at index 0
tuple[0]
```

```python
print(f'tuple1{tuple1}')
```

```python
# list:
# Mutable, can duplicate, itarable,
list1 = ['G','G','Paola','Emil']
list_declared = list(('G','G','Paola','Emil'))
print(f'list1 = {list1}')
print(f'list_declared = {list_declared}')
```

```python
# index notation
print (f'list1 = {list1}')
print(f'first index = {list1[0]}')
print(f'last item in list1 = {list1[-1]}')
print(f'series between 2 & 3 items ={list1[1:3]}') # 2nd item is exclusive
print(f'everything before index 3 = {list1[:3]}')
print(f'everythinh after the first = {list1[1:]}')
```

```python
# list Methods:

## add
list1.append('clare')
print(f'appended clare = {list1}')
## append does not return a list
list2 = list1.append('nathan')
print(f'created list2 = {list2}')
## remove
removed = list1.pop()
print(f'removed {removed}. new list is {list1}')
# change index
list1[1]='J'
print(f'index 1 of list1 changed. new list1 = {list1}')
# combine lists
combined_list1 = [list1, list_declared]
print(f'combined list = {combined_list1}')
# combine list into one
combined_list= [*list1,*list_declared]
print(f'combining list into 1:{combined_list}')

```
