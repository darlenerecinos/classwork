```python
count = 0
while True:
    if count >= 10:
        break # breaks the loop
    count +=1 #  count = count +1
    print(count)
```

```python
# A better option:
count = 0

while count < 10:
    print(count)
    count += 1 
```

```python
is_pass_complete: bool = False


while not is_pass_complete:
    password = input("Enter password in alphabetic characters")
    if not password.isalpha():
        print("Please choose another password with only alphabetic characters")
        continue # restarts / continues the loop from this point... reads nothing below
    is_pass_complete = True
    print("Password Set")
```

```python
from typing import List
# for loop

nums: List[int] = [1,2,3,4,5]

for num in nums: # "num" is variable rename if needed
    num +=1
    if num % 3 == 0: ### % divides number inserted and leaves remainder
         # break would break the loop
        continue #do not print anything divisable by 3
    print(num)
```

```python
sentence: str = "This is a python course."

words: List[str] = sentence.split(" ")
used_vowels = []

for word in words:
    vowels = ["a","e","i","o","u","y"]
    for letter in word:
        if letter in vowels:
            used_vowels.append(letter)
print(used_vowels)

```

```python
# list comprehension, or a generative loop, one-line loop

bar_names = ["Ottoman Empire", "France", "Spain"]

bar_names = [name[:3]+ "." for name in bar_names]

print(bar_names)

#for name in bar_names
   # print(name[:3])
```

```python
# what are functions:
# functions: take an input(argument) proccesses it and thn returns an output(return)
def shorten_names (names_list: List[str], chars: int = 3) -> List[str]:
    """
    Returns a lisr of strings that are the first thrwee characters
    of the word with a period appended.
    
    Args:
        names_list (List[str]): A list of strings that will be shortened
        chars (int): The number of characters to shorten each string to. Must be greater than O. Defaults to 3
        
        Return:
        List[str]: A list of shortened strings to 3 characters
    
    """
    if type(chars) != int:
            raise ValueError ("Chars must be greater than 0")
    if chars < 0:
        raise ValueError ("Chars must be greater than 0")
    names_list = [f"{names[:chars]}." for names in names_list]
    return names_list
```

```python
bar_names = ["Ottoman Empire", "France", "Spain"]

shorten_names (bar_names, chars =2)
```
