```python
# making the year_start better by using a named function
import numpy as np
def get_year_start(row):
    rows = df.loc[
        (df["name"]==row["name"])&
        (df["place"]==row["place"])&
        (df["year"]==row["year"]-1)
    ]
    if len(rows) > 0:
        return np.nan
    date_string = f"{row["year"]}-{choice(range(1,13)):02d}-{choice(range(1,29))}"
    return date_string

df["date_start"]= df.apply(get_year_start, axis=1) # on a df, must pass axis
df.head()

    
```
