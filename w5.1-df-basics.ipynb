```python
## data frame = a collection of series objects
## DF Rules: Columns must be homogenous. All columns must have same numbers of data entries.
## to create a DF: 

# Dictionary creation:
data = {
    "name": ["Nate","Rebecca","Edwin","Preston"],
    "age" : [39,40,11,7],
    "year": ["Senior", "Junior", "Sophmore", "Freshman"]
}
students_df = pd.DataFrame(data=data)
students_df = students_df.set_index("name")
students_df
```

```python
students_df = students_df.reset_index()
students_df
```

```python
# Row-Based creation:
import random

name = ["Messi", "Ronaldo", "Neymar", "Pulisc", "Mbappe"]
goals_per_game = [random.random() for _ in name ]
games = [round(random.random()* 1000) for _ in name ]

cols = ["name", "goals/game", "games"]
stats = [[name[i], goals_per_game[i], games[i]]for i in range(5)]
stats_df = pd.DataFrame(data = stats, columns = cols)
stats_df

```

```python
stats_df["new_col"] = 0
stats_df
```

```python
stats_df = stats_df.drop(columns= ["new_col"])
stats_df
```

```python
stats_df['goals'] = stats_df ['goals/game'] * stats_df ["games"]
stats_df
```

```python
# stats_df = stats_df.rename (columns = {"goals/game": "goals_p_game"})
stats_df= stats_df.rename(columns={"goals/game": "goals_p_game"})
stats_df
```

```python
stats_df["name"] = stats_df ["name"].str.lower()
stats_df
```

```python
## filtering with .loc

stats_df.loc[0]
```

```python
stats_df.loc[0,"name"]
```

```python
stats_df.loc[0]
```

```python
stats_df.loc[:,"name"]
```

```python
# in df, second item after colon is inclusive
stats_df.loc[1:2,"name"]
```

```python
l = [1,2,3,4,5]
l[1:2] # in a list the second item is exclusive
```

```python
stats_df.loc[:,"name":"goals"]
```

```python
# boolean logic with .loc
stats_df.loc[stats_df["goals_p_game"]>0.5]
```

```python
stats_df.loc[stats_df["goals_p_game"]> 0.5, "name"]
```

```python
stats_df.loc[stats_df["name"]== "messi", "name"]= "ronaldo"
stats_df.loc[stats_df["goals_p_game"]> 0.5,"name"] = "messi"
stats_df
```
